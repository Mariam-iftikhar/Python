```python
#Importing pandas and reading excel file.
import pandas as pd
df = pd.read_excel("/content/sample_data/Customer Call List.xlsx")
df
```

# Dropping duplicates from the dataframe.

```python
df=df.drop_duplicates()
df
```

# Dropping column that's not needed.

```python
df=df.drop(columns=['Not_Useful_Column'])
df
```

# Cleaning Last_Name column.

```python
df['Last_Name']=df['Last_Name'].str.strip("/")
df['Last_Name']=df['Last_Name'].str.strip("...")
df['Last_Name']=df['Last_Name'].str.strip("__")
df
```

# Cleaning Phone_Number and converting it to a string.

```python
df['Phone_Number']=df['Phone_Number'].str.replace('[^a-zA-Z0-9]','')
df['Phone_Number']=df['Phone_Number'].apply(lambda x: str(x))
df
```

# Converting Phone_number to proper format and replacing Na values to Null.

```python
df['Phone_Number']=df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number']=df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

# Splitting Address to 3 different columns to store each in those columns.

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',2, expand=True)
df
```

# Replacing Yes/No with Y/N in Do_Not_Contact column.

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

# Replacing NA/ None values with ' '.

```python
df=df.fillna('')
df
```

# Dropping records where Do_Not_Contact is Yes.

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)

df
```

# Dropping records where Phone Number is empty.

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)

df
```

# Reseting the index of the dataframe.

```python
df=df.reset_index(drop=True)
df
```

# Replacing Yes/No with Y/N in Paying Customer column.

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No','N')
df
```
