Loading and Preprocessing:

```python
import pandas as pd
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_excel('/content/AnalysisData (1).xlsx')
df.head()
```

One-Hot Encoding:


```python
df1=pd.get_dummies(df)
df2=df1.iloc[:,1:]
df2
```

Apriori Algorithm:

```python
frequentitems=apriori(df2,min_support=0.07,use_colnames=True)
frequentitems
```

Apriori Algorithm:

```python
rule=association_rules(frequentitems,metric="lift",min_threshold=1)
rule
```

Filtering Rules by Lift:

```python
rule[(rule["lift"]==3) ]
rule
```

Rule 0:

Antecedents: Product 1 name_Apple
Consequents: Product 2 name_Milk
Support: 0.071429
Confidence: 0.25
Lift: 1.75
Interpretation: There is a 25% chance of buying Milk when Apple is purchased, and this association has a lift of 1.75, indicating a positive correlation between buying Apple and Milk.
